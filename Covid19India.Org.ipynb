```python
req = requests.get('https://api.covid19india.org/state_district_wise.json')
```

```python
dat = json.loads(req.text)
dat
```

```python
df = json_normalize(dat)
df.head(3)
```

```python
ab = df.transpose()
```

```python
ab
```

```python
ab['splitdata'] = ab.index
```

```python
ab
```

```python
ab.reset_index(level=0, inplace=True)
```

```python
ab
```

```python
ef = ab.drop(['index'], axis=1)
ef
```

```python
ef[['state','districtdata','district','status','a','c','d','e']] = ef['splitdata'].str.split('.',expand=True)
```

```python
ef
```

```python
qw= ef.drop(['splitdata'], axis = 1)
qw
```

```python
jk=qw.drop(['districtdata'], axis=1)
jk
```

```python
jk.state.nunique()
```

```python
pq=jk.rename(columns={'state':'States and UTs', '0':'Number of Cases', 'a':'A','c':'B','d':'C','e':'D'})
pq
```

```python
lk=pq.rename(columns={0:'Cases'})
lk
```

```python
lk.Cases.unique()
```

```python
lk
```

```python
cols = ['Cases']
mask = lk[cols].applymap(lambda x: isinstance(x, (int, float)))

lk[cols] = lk[cols].where(mask)
print (lk)
```

```python
lk
```

```python
mn = lk.replace('NaN',0)
mn
```

```python
mn.fillna(0, inplace=True)
```

```python
mn
```

```python
rs=mn.groupby(['States and UTs','district','status'])[mn.columns[0]].sum()
rs
```
